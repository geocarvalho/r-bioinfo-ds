# Praticando boxplots

Vamos transportar quartis e boxplots para o universo do R, que está mais do que pronto para isso, portanto não teremos trabalho algum. Criaremos uma lista qualquer de números, como:

```r
numeros <- c (1, 3, 5, 6, 10, 19, 23, 5, 7, 89, 15, 14, 22, 23, 32, 23, 37)
```

Para obtermos o primeiro e terceiro quartis usamos o `summary()`.

```r
summary(numeros)
```

Neste caso, o primeiro quartil é 6, o terceiro é 23. E se quisermos o `boxplot`, usamos o comando abaixo, e ele automaticamente desenha-o para nós. Simples assim, sem nenhum segredo.

```r
boxplot(numeros)
```

É possível customizar esse gráfico, e para isso basta consultar o manual do *boxplot*:

```r
?boxplot
```

Aproveitando, um truque que pode ser muito útil para quando, após gerarmos o gráfico e o boxplot, queremos um arquivo a ser incluído em um documento, como um *paper*. Ou seja, quando queremos salvar isso em uma imagem. Para isso, utilizaremos uma função denominada `png()`. Ela recebe um atributo, o parâmetro `file`, seguido de aspas, para que seja passado o caminho de onde se deve salvar o arquivo.

```r
png(file="/home/genomika/Dados/boxplot.png")
```

Podemos inclusive definir as dimensões da imagem: width (largura) de `700px`, e height (altura) de `700px`.

```r
png(file="/home/genomika/Dados/boxplot.png", width=700, height=700)
```

> Caso digitemos algo errado, podemos usar o atalho "Command + C" (ou "Ctrl + C"). O atalho que traz o último comando digitado é a seta para a cima, no teclado. Se pressionarmos várias vezes, teremos o histórico dos comandos digitados, bastando selecionar um deles seguido de "Enter".

Em seguida, geraremos o *boxplot*.

```r
boxplot(numeros)
```

O programa não faz nada. Por quê? Porque, agora, ele entendeu que não precisa plotar na tela, e sim no arquivo. Para indicar que finalizamos, e que o arquivo pode ser salvo da maneira em que está, utilizamos o seguinte comando:

```r
dev.off()
```

Será retornado `1`, que indica sucesso. Se solicitarmos ao terminal para que sejam listados todos os arquivos (`ls`), teremos o `boxplot.png` que, ao ser aberto com:

```r
open boxplot.png
```
