Essa é a variável `lizard`, com vários números:

```r
lizard = c(6.2,6.6,7.1,7.4,7.6,7.9,8,8.3,8.4,8.5,8.6,8.8,8.8,9.1,9.2,9.4,9.4,9.7,9.9,10.2,10.4,10.8,11.3,11.9)
```

Queremos calcular o intervalo de confiança nessa distribuição, mas o R não nos traz uma função padrão para isso. Sendo assim, usaremos um teste de hipótese que tem isso como saída, o `t.test()` passando lizard. Teremos com `95%` de confiança que o intervalo está entre `8.29` e `9.49`.

Podemos até mudar o nível de confiança, o qual não precisa ser `95%`, e sim `90%`. Obteremos um intervalo ainda mais restrito, de `8.39` até `9.39`.

```r
t.test(lizard, conf.level=0.9)
```

É assim que calculamos intervalos de confiança usando `t.test()` e ignorando o resto. Vamos calcular a média dessa distribuição e desenhar um histograma:

```r
media <- mean(lizard)
hist(lizard)
```

Vamos desenhar um traço azul em cima deste gráfico, na média, com `abline()`, em que `v` significa "valor", e a grossura da linha será `2`.

```r
media <- mean(lizard)
hist(lizard)
abline(v=media, col="blue", lwd=2)
```

Podemos, além disso, incluir uma linha vermelha e mais grossa no primeiro limite do intervalo de confiança, `8.39`.

```r
media <- mean(lizard)
hist(lizard)
abline(v=media, col="blue", lwd=2)
abline(v=8.39, col="red", lwd=4)
```

Feito isso, acrescentaremos outra linha:

```r
media <- mean(lizard)
hist(lizard)
abline(v=media, col="blue", lwd=2)
abline(v=8.39, col="red", lwd=4)
abline(v=9.39, col="red", lwd=4)
```

Lembrando que quando plotávamos e salvávamos imagens em um determinado arquivo, podíamos chamar funções embaixo de funções? Foi assim que chamamos o `hist()`, ele desenhou um histograma, e o `abline()` ficou desenhando linhas.

Tendo a média, observemos o intervalo de confiança, no caso, `90%`.
