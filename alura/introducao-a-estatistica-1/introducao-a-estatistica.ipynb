## Variáveis

1. Categóricas: Não comparáveis (Masculino, feminino, não declarado)
2. Ordinal: Existem ordem, mas sem comparações entre os valores (1 < 2 < 3 < 4 < ... < 10), (altura)
3. Intervalar: A diferença entre cada valor é precisa e mensurável (25ºC, 25.2ºC, 26ºC) 
4. Racional: Parecido com Invelar, sendo 0 a ausência de valor (temperatura de Kelvin).
